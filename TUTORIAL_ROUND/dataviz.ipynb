```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données
day1 = pd.read_csv('data/prices_round_0_day_-1.csv', sep=';')
day2 = pd.read_csv('data/prices_round_0_day_-2.csv', sep=';')

def plot_product(product_name, day1_df, day2_df):
    d1 = day1_df[day1_df['product'] == product_name].copy()
    d2 = day2_df[day2_df['product'] == product_name].copy()

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))

    for ax, d, label in [(ax1, d1, 'Day -1'), (ax2, d2, 'Day -2')]:
        ax.plot(d['timestamp'], d['mid_price'], linewidth=1.5)
        low, high = d['mid_price'].min(), d['mid_price'].max()
        spread = (d['ask_price_1'] - d['bid_price_1']).mean()
        stats = f"Low: {int(low):,}\nHigh: {int(high):,}\nRange: {int(high - low)}\nAvg Spread: {spread:.1f}"
        ax.text(0.98, 0.98, stats, transform=ax.transAxes,
                va='top', ha='right', fontsize=11, family='monospace',
                bbox=dict(boxstyle='round', alpha=0.1))
        ax.set_title(label, fontsize=18)
        ax.set_xlabel('Timestamp')
        ax.set_ylabel('Price')
        ax.grid(True, alpha=0.3)
        ax.ticklabel_format(style='plain', axis='x')

    fig.suptitle(product_name, fontsize=24, fontweight='bold')
    plt.tight_layout()
    plt.close()

    low1 = d1['mid_price'].min(); high1 = d1['mid_price'].max()
    spread1 = (d1['ask_price_1'] - d1['bid_price_1']).mean()
    low2 = d2['mid_price'].min(); high2 = d2['mid_price'].max()
    spread2 = (d2['ask_price_1'] - d2['bid_price_1']).mean()

    print(f"\n{product_name} - Résumé:")
    print(f"Day -1: Low={int(low1)}, High={int(high1)}, Range={int(high1-low1)}, AvgSpread={spread1:.1f}")
    print(f"Day -2: Low={int(low2)}, High={int(high2)}, Range={int(high2-low2)}, AvgSpread={spread2:.1f}")

# Visualiser les deux produits
plot_product('TOMATOES', day1, day2)
plot_product('EMERALDS', day1, day2)
```
