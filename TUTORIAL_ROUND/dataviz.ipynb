```python
import pandas as pd
import matplotlib.pyplot as plt

# Configuration du style
plt.rcParams['figure.facecolor'] = '#1e2530'
plt.rcParams['axes.facecolor'] = '#1e2530'
plt.rcParams['axes.edgecolor'] = '#3d4a5c'
plt.rcParams['grid.color'] = '#2d3847'
plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['xtick.color'] = 'white'
plt.rcParams['ytick.color'] = 'white'

# Charger les données
day1 = pd.read_csv('data/prices_round_0_day_-1.csv', sep=';')
day2 = pd.read_csv('data/prices_round_0_day_-2.csv', sep=';')

def plot_product(product_name, day1_df, day2_df):
    """Trace le produit pour les 2 jours côte à côte"""
    
    # Filtrer pour le produit
    d1 = day1_df[day1_df['product'] == product_name].copy()
    d2 = day2_df[day2_df['product'] == product_name].copy()
    
    # Créer la figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6), facecolor='#1e2530')
    
    # Day 1
    ax1.plot(d1['timestamp'], d1['mid_price'], color='#00ff88', linewidth=1.5)
    low1 = d1['mid_price'].min()
    high1 = d1['mid_price'].max()
    range1 = high1 - low1
    d1['spread'] = d1['ask_price_1'] - d1['bid_price_1']
    avg_spread1 = d1['spread'].mean()
    
    stats1 = f"Low: {int(low1):,}\nHigh: {int(high1):,}\nRange: {int(range1)}\nAvg Spread: {avg_spread1:.1f}"
    ax1.text(0.98, 0.98, stats1, transform=ax1.transAxes, 
             va='top', ha='right', fontsize=11, family='monospace',
             bbox=dict(boxstyle='round', facecolor='#2d3847', alpha=0.8, edgecolor='#00ff88'),
             color='white')
    
    ax1.set_title('Day -1', color='white', fontsize=18, pad=20)
    ax1.set_xlabel('Timestamp', color='white')
    ax1.set_ylabel('Price', color='white')
    ax1.grid(True, alpha=0.2)
    ax1.ticklabel_format(style='plain', axis='x')
    
    # Day 2
    ax2.plot(d2['timestamp'], d2['mid_price'], color='#00ff88', linewidth=1.5)
    low2 = d2['mid_price'].min()
    high2 = d2['mid_price'].max()
    range2 = high2 - low2
    d2['spread'] = d2['ask_price_1'] - d2['bid_price_1']
    avg_spread2 = d2['spread'].mean()
    
    stats2 = f"Low: {int(low2):,}\nHigh: {int(high2):,}\nRange: {int(range2)}\nAvg Spread: {avg_spread2:.1f}"
    ax2.text(0.98, 0.98, stats2, transform=ax2.transAxes,
             va='top', ha='right', fontsize=11, family='monospace',
             bbox=dict(boxstyle='round', facecolor='#2d3847', alpha=0.8, edgecolor='#00ff88'),
             color='white')
    
    ax2.set_title('Day -2', color='white', fontsize=18, pad=20)
    ax2.set_xlabel('Timestamp', color='white')
    ax2.set_ylabel('Price', color='white')
    ax2.grid(True, alpha=0.2)
    ax2.ticklabel_format(style='plain', axis='x')
    
    # Titre général
    fig.suptitle(product_name, color='#00ff88', fontsize=24, fontweight='bold', y=0.98)
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig(f'{product_name}.png', dpi=150, facecolor='#1e2530')
    plt.show()
    
    print(f"\n{product_name} - Résumé:")
    print(f"Day -1: Low={int(low1)}, High={int(high1)}, Range={int(range1)}, AvgSpread={avg_spread1:.1f}")
    print(f"Day -2: Low={int(low2)}, High={int(high2)}, Range={int(range2)}, AvgSpread={avg_spread2:.1f}")

# Visualiser les deux produits
plot_product('TOMATOES', day1, day2)
plot_product('EMERALDS', day1, day2)
```
